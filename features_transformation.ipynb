# Feature Transformation
- converting skewed data to Normal Distributed data

```python
df[continuous].skew()
```

***avg_glucose_level is right skewed***

```python
# log transformation- converting each value to its log value

df['log_avg_glucose_level']=np.log(df['avg_glucose_level']) # input as 1 dimensional array
df['log_avg_glucose_level']
```

```python
df['log_avg_glucose_level'].skew()
```

```python
df
```

***Root transformation***
- nth root

```python
df['root_avg_glucose_level']=df['avg_glucose_level']**(1/7)
df['root_avg_glucose_level'].skew()
```

```python
df
```

***BoxCox***

```python
from scipy.stats import boxcox
boxcox(df['avg_glucose_level'])
```

```python
df['boxcox_avg_glucose_level'], param = boxcox(df['avg_glucose_level'])
print(df['boxcox_avg_glucose_level'])

print(f'\nlambda value is : {param}')
```
